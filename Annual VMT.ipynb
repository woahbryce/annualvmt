```python
#Import libraries
import pandas as pd
import geopandas as gpd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df_fp = r"C:\Users\Bryce\Google Drive\Research\Annual VMT by State\data.csv"
shp_fp = r"C:\Users\Bryce\Google Drive\Research\Resources\ne_110m_admin_1_states_provinces"

df = pd.read_csv(df_fp)
df.head()
```

```python
state_shp = gpd.read_file(shp_fp, encoding='utf-8')
state_shp.head()
```

```python
df.columns = map(str.lower, df.columns)
df.head()
```

```python
choro_merged = pd.merge(state_shp, df, how='left', left_on='name', right_on='state')
choro_merged
```

```python
sns.distplot(choro_merged['vehicle-miles per licensed driver'])
```

```python
variable = 'vehicle-miles per licensed driver'
vmin, vmax = min(choro_merged[variable]), max(choro_merged[variable]) 
color = 'Oranges'
fig, ax = plt.subplots(figsize=(18,10))
choro_merged.plot(column=variable, cmap=color, linewidth=1.2, ax=ax, edgecolor='0.8')
ax.axis('off')
ax.set_title('Annual Vehicle-Miles Traveled by Licensed Drivers', fontdict={'fontsize':'18', 'fontweight':'3'})
ax.annotate('Source: FHWA, 2000', xy=(0.1, .08), xycoords='figure fraction', horizontalalignment='left',
           verticalalignment='top', fontsize=12, color='#555555')
#Colorbar Legend
sm = plt.cm.ScalarMappable(cmap=color, norm=plt.Normalize(vmin=vmin, vmax=vmax))
# Empty array for data range
sm._A = []
# Add colorbar to figure
cbar = fig.colorbar(sm)
# Despine colorbar
cbar.ax.tick_params(right=False)
```
